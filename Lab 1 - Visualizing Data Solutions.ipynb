# Visualizing Data

Python has many tools available for visualizing data. In this class, we will use the graphics libraries matplotlib and Seaborn. First, we need to import these libraries. We will also use the command `%matplotlib inline` to make graphics visible. In future labs, we will run these commands at the start of every worksheet.

````python
#Evaluate this cell
%matplotlib inline
import matplotlib as mat
import seaborn as sns
````

Probably the simplest way to display quantitative data is just to make each data point a dot. This produces what is known as a *dotplot*.

Using Seaborn, we can use the `stripplot` function to make such a plot. 

Example: `sns.stripplot(data=datalist, orient="vertical");`


````python
#6 Create a dotplot of the LA rainfall data
sns.stripplot(data=rain, orient="vertical")
````

When displaying data, we almost always want to label the axes. To do this with Seaborn, we first assign our graph to a variable, say p. Then, we can use the p.set function to set an x label or a y label, depending on how the plot is oriented. Many Seaborn plotting functions use the same syntax.

Example: `p=sns.stripplot(data=datalist, orient="vertical")`

`p.set(ylabel="Measurements (units)")`

````python
#7. Copy your plotting code from the previous exercise and label your axes.
p=sns.stripplot(data=rain, orient="vertical")
p.set(ylabel="Units of Rain")
````

From now on, you should always appropriately label your axes.

Another, way to plot points, which sometimes looks neater, is to use what is called a beeswarm plot. This spaces the points out systematically rather than randomly. You can do this in Seaborn with the `swarmplot` function.

Example: `sns.swarmplot(data=datalist)`

````python
#8. Make a beeswarm plot of the rainfall data.
p=sns.swarmplot(data=rain)
p.set(xlabel="Units of Rain")
````

While dot plots show every data point individually, sometimes we get more insight from combining values into bins and showing how many observations are in each bin. This kind of plot is called a histogram. 

To make a histogram with Seaborn, use the function `displot`.

Example:  
```
p=sns.displot(data=datalist);

p.set(xlabel="Label")
```

````python
#9. Make a histogram of the rainfall data. Label the x axis.
p=sns.displot(data=rain)
p.set(xlabel="Units of Rain")
````

The shape of a histogram is strongly affected by the number of bins used in making it. Usually, the default chosen by Seaborn works pretty well, but you should look at what happens when you change the number of bins in your histogram. This is done with the option `bins=n`, where `n` is the number of bins you want to break up your data into.

````python
#10. Try at least two different numbers of bins for the rainfall data histogram.
p=sns.displot(data=rain, bins=4)
p.set(xlabel="Units of Rain")
````

````python
p=sns.displot(data=rain, bins=10)
p.set(xlabel="Units of Rain")
````

You can also use `displot` to create a *kernel density estimator (KDE)* plot. A KDE plot is basically a smoothed histogram. You will learn more about how these plots work in lecture and the readings. To overlay one on a histogram, just use the plotting option `kde=True`.

````python
#11. Display the histogram you made in #9 with the KDE.
p=sns.displot(data=rain, kde=True)
p.set(xlabel="Units of Rain")
````

We can also put two KDEs together to make a violin plot using `sns.violinplot(data=datalist)`.

````python
#12. Make a violin plot of the rainfall data
p=sns.violinplot(data=rain)
p.set(ylabel="Units of rain")
````

For larger datasets, it can be useful to display summary statistics. A boxplot shows the median of a dataset, its 25th and 75th percentiles, and information about outliers (data points very different from most others).

Example: 
`p=sns.boxplot(data=datalist, color="green", orient="vertical");`

`p.set(ylabel="Units of rain")`

````python
#13. Make a box plot of the rainfall data. Make it whatever color you want.
#HINT: Axes labels for boxplots (and for most figures) work the same way they do for dotplots.
p = sns.boxplot(data=rain, color="red", orient="vertical")
p.set(ylabel="Units of rain")
````
